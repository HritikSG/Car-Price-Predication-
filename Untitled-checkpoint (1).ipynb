```python
import pandas as pd

data = pd.read_csv("cars.csv")
data.head()
```

```python
# filter the data
fdata = data[["Year", "Engine HP", "Engine Cylinders", "Number of Doors", "Popularity", "MSRP"]]
fdata.shape
```

```python
fdata = fdata.rename(columns = {"Year": "year", "Engine HP": "hp", "Engine Cylinders": "cylinders", 
              "Number of Doors": "ndoors", "Popularity": "popularity", "MSRP": "price"})
fdata.head()
```

```python
import matplotlib.pyplot as plt

prices = fdata[["year", "price"]].groupby("year").mean()["price"]
plt.plot(prices)
```

```python
# def zero_center(column: pd.Series):
#     mean = column.mean()
#     return (column - mean).values

def max_scaling(column: pd.Series):
    maxi = max(column)
    return column / maxi


# def standard_scaling(column: pd.Series):
#     mean_value = column.mean()
#     std_value = column.std()
#     standardized_column = (column - mean_value) / std_value
#     return standardized_column

# from sklearn.preprocessing import QuantileTransformer

# def gaussian_transformation(column: pd.Series):
    
#     # Reshape the data as QuantileTransformer expects a 2D array
#     reshaped_column = column.values.reshape(-1, 1)

#     # Initialize the transformer
#     transformer = QuantileTransformer(output_distribution='normal')

#     # Fit the transformer and transform the data
#     transformed_column = transformer.fit_transform(reshaped_column)

#     # Reshape the transformed data back to a 1D array
#     transformed_column = transformed_column.flatten()

#     return transformed_column, transformer

# # Example usage
# column_name = 'year'  # Replace with the actual column name in your dataset
# column = fdata[column_name]
# norm_data, transformer = gaussian_transformation(column)
```

```python
norm_data = []

for column_name in fdata:
    print(column_name)
    column = fdata[column_name]
    norm_data.append(max_scaling(column))
```

```python
data.shape[0]
```

```python
# norm_data
```

```python
import numpy as np

X = []
Y = []

for i in range(data.shape[0]):
    features = []
    features.append(norm_data[0][i])
    features.append(norm_data[1][i])
    features.append(norm_data[2][i])
    features.append(norm_data[3][i])
    features.append(norm_data[4][i])
    X.append(features)
    Y.append(norm_data[5][i])

    
X = np.array(X)
Y = np.array(Y)
```

```python
X = X.reshape((-1, 1, 5))
Y = Y.reshape((-1, 1, 1))
```

```python
from sklearn.model_selection import train_test_split

# train, validation, test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
X_val, X_test, Y_val, Y_test = train_test_split(X_test, Y_test, test_size=0.2) 

print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
```

```python
import tensorflow as tf
tf.random.set_seed(42)

model = tf.keras.models.Sequential([
    tf.keras.layers.Input((1, 5)),
    tf.keras.layers.Dense(5),
    tf.keras.layers.Dense(3),
    tf.keras.layers.Dense(1, activation = "tanh")
], name = "ANN")

model.compile(loss="mse", optimizer="adam")
model.summary()
```

```python
history = model.fit(X_train, Y_train, epochs = 5, validation_data=(X_val, Y_val))
```

```python
plt.plot(history.history["loss"], label = "Training loss")
plt.plot(history.history["val_loss"], label = "Validation loss")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

y_hat = model.predict(X_test)
mse = mean_squared_error(Y_test.ravel(), y_hat.ravel())
print("Testing MSE:", mse)
```

```python
actual_diff = round(mse * max(fdata["price"]), 2)
print("Actual difference in predictions: $", actual_diff)
```

```python
import numpy as np

# Assuming y_hat and Y_test are numpy arrays
predicted_prices = y_hat.ravel()
actual_prices = Y_test.ravel()

# Display the predicted and actual prices
for i in range(len(predicted_prices)):
    print(f"Actual Price: {actual_prices[i]}, Predicted Price: {predicted_prices[i]}")
```

```python
maxi = max(fdata['price'])
rnd = lambda x: round(x, 2)

ac = np.array(list(map(rnd, actual_prices * maxi)))
pr = np.array(list(map(rnd, predicted_prices * maxi)))
```

```python
pd.set_option("display.float_format", "{:.2f}".format)
```

```python
se = np.array(list(map(lambda x: rnd(np.abs(x)), (ac - pr))))

data = {"Acutal price $": ac, "Predicted price $": pr, "Standard Error": se}
data = pd.DataFrame(data)
data.head(10)
```

```python
plt.figure(figsize=(14, 5))

plt.hist(data["Standard Error"], bins = 25)
plt.title("Error distribution")
plt.xlabel("Car price $")
plt.ylabel("Number of samples")
plt.xticks(range(0, 50000, 5000))
plt.show()
```

```python
good = data[data["Standard Error"] <= data["Standard Error"].mean()].__len__()
bad = data[data["Standard Error"] > data["Standard Error"].mean()].__len__()

print(good, bad)
```
